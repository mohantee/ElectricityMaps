# API Exploration
Explore the Electricity Maps API schemas and save fixtures for testing.

```python
import json
from datetime import UTC, datetime, timedelta

from electricity_maps.api.client import ElectricityMapsClient
```

```python
client = ElectricityMapsClient()
end = datetime.now(UTC)
start = end - timedelta(hours=3)
```

```python
print(f"Fetching mix data for FR from {start} to {end}")
mix_resp = client.get_raw_mix_range("FR", start, end)
print(f"Mix records: {len(mix_resp['data'])}")
print("Example Mix:")
print(json.dumps(mix_resp['data'][0], indent=2))
```

```python
print(f"Fetching flows data for FR from {start} to {end}")
flows_resp = client.get_raw_flows_range("FR", start, end)
print(f"Flows records: {len(flows_resp['data'])}")
print("Example Flows:")
print(json.dumps(flows_resp['data'][0], indent=2))
```
